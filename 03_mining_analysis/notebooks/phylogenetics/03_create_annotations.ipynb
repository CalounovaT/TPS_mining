# Color characterized leaves by the type

```python
def generate_color_annotation(leaves_file, colors_file, output_file):
    # Read leaves and their assigned classes
    with open(leaves_file, 'r') as leaves_file:
        leaves_data = [line.strip().split('\t') for line in leaves_file.readlines()]

    # Read colors assigned to classes
    with open(colors_file, 'r') as colors_file:
        colors_data = [line.strip().split() for line in colors_file.readlines()]

    # Create a dictionary to map classes to colors
    class_color_mapping = {class_id: color for color, class_id in colors_data}

    # Generate color annotation text
    color_annotation_text = 'TREE_COLORS\n'
    color_annotation_text += 'SEPARATOR SPACE\n'
    color_annotation_text += 'DATA\n'

    color_annotation_text += "#NODE_ID TYPE COLOR LABEL_OR_STYLE SIZE_FACTOR\n"

    for (leaf_id, class_id) in leaves_data:
        #print(leaf_id, class_id)
        if class_id in class_color_mapping:
            color_annotation_text += f"{leaf_id} range {class_color_mapping[class_id]} bold 1\n"

    # Write the color annotation text to the output file
    with open(output_file, 'w') as output_file:
        output_file.write(color_annotation_text)
```

```python
def generate_color_annotation2(leaves_file, colors_file, output_file):
    # Read leaves and their assigned classes
    with open(leaves_file, 'r') as leaves_file:
        leaves_data = [line.strip().split('\t') for line in leaves_file.readlines()]

    # Read colors assigned to classes
    with open(colors_file, 'r') as colors_file:
        colors_data = [line.strip().split() for line in colors_file.readlines()]

    # Create a dictionary to map classes to colors
    class_color_mapping = {class_id: color for color, class_id in colors_data}

    # Generate color annotation text
    color_annotation_text = 'DATASET_SYMBOL\n'
    color_annotation_text += 'SEPARATOR TAB\n'
    color_annotation_text += 'DATASET_LABEL\ttype\n'
    color_annotation_text += 'DATA\n'

    color_annotation_text += "#ID\tsymbol\tsize\tcolor\tfill\tposition\tlabel\n"

    for (leaf_id, class_id) in leaves_data:
        #print(leaf_id, class_id)
        if class_id in class_color_mapping:
            color_annotation_text += f"{leaf_id}\t2\t10\t{class_color_mapping[class_id]}\t1\t1\n"

    # Write the color annotation text to the output file
    with open(output_file, 'w') as output_file:
        output_file.write(color_annotation_text)
```

```python
def collapse_unchar(tree, characterized_leaves_ids, out_file):
    with open(out_file, 'w') as out_handle:
        out_handle.write('COLLAPSE'+'\n')
        out_handle.write('DATA'+'\n')
        for node in tree.get_descendants('preorder'):
            node_leaves = node.get_leaves()
            node_leaves = {l.name for l in node_leaves}
            intersection = node_leaves.intersection(characterized_leaves_ids)
            if len(intersection)==0:
                if len(node_leaves)>1:
                    out_handle.write(("|".join(node_leaves))+'\n')
```

```python
def collapse_char(tree, types_dict, characterized_leaves_ids, out_file):
    with open(out_file, 'w') as out_handle:
        out_handle.write('COLLAPSE'+'\n')
        out_handle.write('DATA'+'\n')
        for node in tree.get_descendants('preorder'):
            node_leaves = node.get_leaves()
            node_leaves = {l.name for l in node_leaves}
            char = {char_info_dict[leaf]['characterized'] if leaf in char_info_dict.keys() else 0 for leaf in node_leaves}
            if char == {1}:
                node_types = {types_dict[name]['type'] for name in node_leaves}
                if len(node_types)<=2:
                    if len(node_leaves)>1:
                        if 'mixed' in node_types:
                            out_handle.write(("|".join(node_leaves))+'\n')
                        elif len(node_types)==1:
                            out_handle.write(("|".join(node_leaves))+'\n')
```

```python
def get_collapsed_clade_domain(domain, threshold=0.5):
    domain_collapesed_clades = []
    with open('annotations/collapsed_unchar.txt','r') as in_handle:
        data_started=False
        for line in in_handle.readlines()[2:]:
            leaves = line.strip().split('|')
            domain_values = []
            for leaf in leaves:
                if leaf in tps_anno_df.index:
                    domain_values.append(tps_anno_df.loc[leaf][domain])
                elif leaf in anno_df.index:
                    domain_values.append(anno_df.loc[leaf][domain])
                else:
                    print(leaf)
            if sum(domain_values)/len(leaves) > threshold:
                domain_collapesed_clades.append(line.strip())
    return domain_collapesed_clades
```
