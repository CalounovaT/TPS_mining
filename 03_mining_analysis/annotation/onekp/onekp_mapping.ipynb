```python
def get_tax_id(x):
    try:
        t = taxoniq.Taxon(scientific_name=x['organism'])
        return t.tax_id
    except:
        if x['organism'] in missing_map.keys():
            return missing_map[x['organism']]
        if isnan(x['Tax_ID']):
            return 'error'
        else:
            return x['Tax_ID']
```

```python
anno_df['tax_id'] = anno_df.apply(lambda x: get_tax_id(x), axis=1)
```

```python
anno_df[anno_df['tax_id'] == 'error']['organism'].value_counts()
```

```python
taxid_map = {323182.0: 2834071, 263337.0: 1630381, 23810.0: 2768810}
```

```python
def change_tax_ids(x):
    if x in taxid_map.keys():
        return taxid_map[x]
    else:
        return x
```

```python
anno_df['tax_id'] = anno_df['tax_id'].apply(lambda x: change_tax_ids(x))
```

```python
def add_lineage(x):
    try:
        t = taxoniq.Taxon(x)
        lineage = dict()
        for line in t.ranked_lineage:
            lineage[line.rank.name] = line.scientific_name
        return lineage
    except:
        return 'error'
```

```python
anno_df['lineage'] = anno_df['tax_id'].apply(lambda x: add_lineage(x))
```

```python
def add_line(lineage, line):
    try:
        return lineage[line]
    except:
        return 'unknown'
```

```python
anno_df['superkingdom'] = anno_df['lineage'].apply(lambda x: add_line(x, 'superkingdom'))
```

```python
anno_df['kingdom'] = anno_df['lineage'].apply(lambda x: add_line(x, 'kingdom'))
```

```python
anno_df['phylum'] = anno_df['lineage'].apply(lambda x: add_line(x, 'phylum'))
```
